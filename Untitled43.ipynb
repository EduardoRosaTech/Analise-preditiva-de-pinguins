```python
import sklearn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados dos pinguins
penguim = sns.load_dataset('penguins')

# Visualizar as primeiras 25 linhas do dataset
penguim.head(25)
```

```python
with sns.axes_style('whitegrid'):
    grafico = sns.pairplot(data=penguim.drop(['sex', 'island'], axis=1), hue="species", palette="pastel")
```

```python
with sns.axes_style('whitegrid'):
    grafico = sns.countplot(data=penguim, x='sex', hue="species", palette="pastel")
```

```python
with sns.axes_style('whitegrid'):
    grafico = sns.countplot(data=penguim, x='island', hue="species", palette="pastel")
```

```python
# Verificando valores nulos
penguim.isnull().sum()

# Preenchendo valores nulos com a média para variáveis numéricas e o valor mais frequente para categóricas
penguim['bill_length_mm'].fillna(penguim['bill_length_mm'].mean(), inplace=True)
penguim['bill_depth_mm'].fillna(penguim['bill_depth_mm'].mean(), inplace=True)
penguim['flipper_length_mm'].fillna(penguim['flipper_length_mm'].mean(), inplace=True)
penguim['body_mass_g'].fillna(penguim['body_mass_g'].mean(), inplace=True)
penguim['sex'].fillna(penguim['sex'].mode()[0], inplace=True)

penguim.isnull().sum()
```

```python
# Transformação de variáveis categóricas
penguim['island_nom'] = penguim['island'].astype('category').cat.codes
penguim['sex_nom'] = penguim['sex'].astype('category').cat.codes

penguim.head()
```

```python
# Reordenando as colunas
penguim = penguim[['species', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'island_nom', 'sex_nom']]
penguim.head()
```

```python
from sklearn.model_selection import train_test_split

# Separação em treino e teste
X = penguim.drop('species', axis=1)
y = penguim['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Treinando o modelo
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Plotando a árvore de decisão
plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=X.columns, class_names=model.classes_)
plt.show()

# Número de folhas
n_leaves = model.get_n_leaves()
print(f"Número de folhas: {n_leaves}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = model.predict(X_test)

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Acurácia
accuracy = accuracy_score(y_test, model.predict(X_test))
print(f"Acurácia: {accuracy:.2%}")
```
